```python
#4. Create a function get_max that takes a list of numbers as a parameter and returns the largest number.
def get_max(numbers):
    max = numbers[0]
    for i in numbers:
        if i > max:
            max = i
    return max
```

```python
get_max([1,9,2,45,7])
```

```python
#8. Use the filter() function with a lambda to get all even numbers from a list [1, 2, 3, 4, 5, 6].

numbers = [1, 2, 3, 4, 5, 6]

even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)
```

```python
#9. Use the map() function with a lambda to convert a list of temperatures in Celsius [0, 10, 20] to Fahrenheit (F = C * 9/5 + 32).

celsius = [0, 10, 20]
fahrenheit = list(map(lambda c: c * 9/5 + 32, celsius))
print(fahrenheit)
```

```python
#10. Write a recursive function countdown(n) that prints numbers from n to 1.

def countdown(n):
    if n <= 0:       
        return
    print(n)
    countdown(n - 1) 

countdown(5)
```
