```python
#1. Convert radians into degrees
import math
def radian(a):
    degree=a*180/math.pi
    return degree
radian(5)
```

```python
# 2. Sort a list
def f(numbers, order):
    if order == 'asc':
        return sorted(numbers)
    elif order == 'desc':
        return sorted(numbers, reverse=True)
    elif order == 'none':
        return numbers
    else:
        return 'Error'
```

```python
f([4,5,8],'desc')
```

```python
#3. Convert a decimal number into binary
def conversion(n):
    if n == 0:
        return "0"
    binary = ""
    while n > 0:
        binary = str(n % 2) + binary
        n = n // 2
    return binary

conversion(16)
```

```python
#4. Count the vowels in a string
def counting(word):
    numbofvoy = 0
    for i in word.lower():
        if i == 'a' or i == 'e' or i == 'i' or i == 'o' or i == 'u':
            numbofvoy += 1
    return numbofvoy

counting('Bertrand')
```

```python
#5. Hide the credit card number
def hiding(cardnumber):
    card_str = str(cardnumber)
    newnumber = "*" * (len(card_str) - 4) + card_str[-4:]
    return newnumber
hiding(1234098779171234)
```

```python
#6. Are the Xs equal to the Os?
def counter(word):
    numbofo=0
    numbofx=0
    for i in word.lower():
        if i == 'o':
            numbofo += 1
        if i == 'x':
            numbofx += 1
    if numbofo==numbofx:
        print (True)
    else:
        print (False)

counter('Saxo')
```

```python
#7. Create a calculator function
def calculate(a,operator,b):
    if operator == '+':
        return a+b
    elif operator == '-':
        return a-b
    elif operator == '*':
        return a*b
    elif operator == '/':
        if b==0 :
            return 'error'
        return a/b
    else:
        return 'Invalid'

calculate(5,'+',4)
```

```python
#8. Give me the discount
def apply(price,discount):
    prix=int(price)
    pourcentage=int(discount)
    newprice=0
    newprice=prix-((prix*pourcentage)/100)
    return newprice

apply(50,5)
```

```python
#9. Just the numbers
def function(list):
    outpout = []
    for i in list:
        if isinstance(i, int):
            outpout.append(i)
    return outpout
function([0,2,'frere'])
```

```python
#10. Repeat the characters
def doubler(word):
    doubledword=''
    for i in word:
        doubledword=doubledword + i*2
    return doubledword
doubler('bonjour')
```

```python
#11. Convert lowercase to uppercase
def uppercase(word):
    return word.upper()

uppercase('bonjour')
```

```python
#12. Add dots
def adding(word):
    return ".".join(word)
adding('bonjour')
```
